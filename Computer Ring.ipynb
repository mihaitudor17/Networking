```python
"""Definim clasa token, ea fiind globala alaturi de lista de calculatoare"""
import threading
import random
import time
count=-1
class token:
    buffer=""
    ipSource=""
    ipDest=""
    busy=False
    destReach=False
tok=token()
compList=[]
"""Generare de IP random"""
def rand_ip():
    ip=""
    for i in range(4):
        temp=random.randint(0,255)
        ip+=str(temp)
        ip+="."
    return ip[:-1]
"""Moves the token"""
def moveTok(index,sens):
    if index>=len(compList):
             index=0
    while compList[index].ip!=tok.ipDest:
        print(compList[index].name+" Mut jetonul")
        index+=1
        if index>=len(compList):
             index=0
    if sens==1:
        print (compList[index].name +" Am ajuns la destinatie")
        compList[index].buffer.append(tok.buffer)
    else:
        print (compList[index].name +" Am ajuns inapoi")
        for i in range (len(compList)):
            print(compList[i].name,end=" ")
            print(compList[i].ip,end=" ")
            if compList[i].buffer:
                print(compList[i].buffer)
            else:
                print("null")
"""Functia care verifica daca s-a ajuns al calculatorul destinatie, si alege o noua destinatie"""
def transfer(comp):
    tok.ipSource=comp.ip
    temp=random.randint(0,9)
    while tok.ipSource==compList[temp].ip:
        temp=random.randint(0,9)
    tok.ipDest=compList[temp].ip
    message="mesaj"
    tok.buffer=message
    print(comp.name+" Am preluat jetonul")
    index=compList.index(comp)
    index+=1
    if index>=len(compList):
             index=0
    moveTok(index,1)
    tok.ipSource,tok.ipDest=tok.ipDest,tok.ipSource
    index=temp+1
    if index>=len(compList):
             index=0
    moveTok(index+1,-1)
"""Clasa computer"""
class computer(threading.Thread):
    def __init__(self,name,ip):
        threading.Thread.__init__(self)
        self.lock = threading.Lock()
        self.name=name 
        self.ip=ip
        self.buffer=[] 
    def run(self):
        while True:
            if tok.busy==False:
                self.lock.acquire()
                tok.busy=True
                transfer(self)
                tok.busy=False
                self.lock.release()
                time.sleep(0.1)
            else:
                time.sleep(0.5)
"""Functia main"""
def main():
    count=int(input("Introduceti numarul de cicluri: "))
    for i in range(10):
        temp=computer("Computer"+str(i+1), rand_ip())
        compList.append(temp)
    for i in range(10):
        compList[i].start()
if __name__ == "__main__":
    main()
```
