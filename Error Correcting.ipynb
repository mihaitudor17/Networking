```python
#Implementare algoritm CRC

def read():
    message=input("Introduceti mesajul: ")
    message+=" "
    message+=input("Introduceti generatorul: ") 
    return message
def CRC(message):
    temp=message.split()
    extended=temp[0]
    generator=temp[1]
    all_binary = all(c in '01' for c in temp[0])
    all_binary1= all(c in '01' for c in temp[1])
    if all_binary==False or all_binary1==False:
        return "Nu ati introdus biti"
    extended+="0"*(len(generator)-1)
    temp.append(extended)
    while len(extended)>=len(generator):
        i=0
        copy=[]
        while i<len(generator):
            if extended[i]==generator[i]:
                copy.append('0')
            else:
                copy.append('1')
            i=i+1
        while i<len(extended):
            copy.append(extended[i])
            i=i+1
        extended=copy
        while len(extended)>=len(generator) and extended[0]!='1':
            extended=extended[1:]
    result=extended
    extended=temp[2]
    i=len(extended)-len(result)
    j=0
    copy=extended[:i]
    while i<len(extended):
        if extended[i]==result[j]:
            copy+='0'
        else:
            copy+='1'
        i+=1
        j+=1
    final=copy
    return final
def main():
    print("Mesajul final este: "+str(CRC(read())))
if __name__ == "__main__":
    main()
```

```python
#Implementare biti de paritate bidimensionala
import random
def read():
    message=input("Introduceti mesajul: ")
    message=stringToByte(message)
    return message

def stringToByte(message):
    binary=""
    message=message.replace(" ","")
    for x in message:
        num=bin(ord(x)).replace("0b","")
        temp="0"*(7-len(num))
        temp+=num
        binary+=temp
    return binary
     
def PB(message):
    i=0
    row=[]
    if len(message)%7 != 0:
        return "Numarul de cifre trebuie sa fie divizibil cu 7"
    while i<len(message):
        row.append(message[i:i+7])
        i+=7
    column=""
    for x in row:
        suma=0
        for i in x:
            if i=='1':
                suma+=1
        column+=str(suma%2)
    final=""
    for i in range(7):
        s=0
        for x in row:
             if x[i] == '1':
                s+=1
        if s%2==0:
            final+='0'
        else:
            final+='1'
    suma=0
    for x in column:
        suma+=int(x)
    if suma%2==0:
        final+='0'
    else:
        final+='1'
    suma+=int(final[:-1:])
    column+=str(suma%2)
    return final+" "+column
def repair(words1,words2,messageWrong):
    x=-1
    y=-1
    for i in range(0,len(words1[0])-1):
        if words1[0][i]!=words2[0][i]:
            x=i
            i=len(words1[0])
    for i in range(0,len(words1[1])-1):
        if words1[1][i]!=words2[1][i]:
            y=i
            i=len(words1[1])
    temp=((int(messageWrong[7*y+x])+1)%2)
    return messageWrong[:7*y+x]+str(temp)+messageWrong[7*y+x+1:]
           
    
def main():
    message=read()
    words=PB(message).split()
    num=random.randint(0,len(message)-1)
    temp=str((int(message[num])+1)%2)
    messageWrong=message[:num]+temp+message[num+1:]
    wordsWrong=PB(messageWrong).split()
    if words==wordsWrong:
        print("Mesajul este corect: "+message)
    else:
        print("Mesaj gresit:   "+str(messageWrong))
        print("Mesaj corectat: "+repair(words,wordsWrong,messageWrong))
if __name__ == "__main__":
    main()
```
